```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python

def draw_screen(ax, title, elements):
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 15)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(title, fontsize=12)

    for element in elements:
        type, position, size, text = element
        if type == "box":
            ax.add_patch(patches.Rectangle(position, size, fill=False))
            ax.text(position[0] + 0.1, position[1] + size[1] / 2, text, verticalalignment='center', fontsize=10)
        elif type == "text":
            ax.text(position[0], position[1], text, verticalalignment='center', fontsize=8)

fig, axs = plt.subplots(3, 3, figsize=(15, 15))

screens = [
    ("Notification Screen", [
        ("box", (1, 12), (8, 2), "New Survey Available"),
        ("text", (1, 14), "Notification appears in the Android Notification Bar.")
    ]),
    ("Survey Start Screen", [
        ("box", (1, 10), (8, 4), "Welcome! Start Survey"),
        ("text", (1, 14), "User opens the app and sees a welcome message with a 'Start Survey' button.")
    ]),
    ("Question 1 (Efficiency)", [
        ("box", (1, 10), (8, 4), "Was the lecture content delivered efficiently?\n1  2  3  4  5"),
        ("text", (1, 14), "'Was the lecture content delivered efficiently?' with a Likert scale (1-5).")
    ]),
    ("Question 2 (Clarity)", [
        ("box", (1, 10), (8, 4), "Was the lecture material clear and understandable?\n1  2  3  4  5"),
        ("text", (1, 14), "'Was the lecture material clear and understandable?' with a Likert scale (1-5).")
    ]),
    ("Question 3 (Stimulation)", [
        ("box", (1, 10), (8, 4), "Did you find the lecture content stimulating?\n1  2  3  4  5"),
        ("text", (1, 14), "'Did you find the lecture content stimulating?' with a Likert scale (1-5).")
    ]),
    ("Question 4 (Dependability)", [
        ("box", (1, 10), (8, 4), "Was the lecture structure predictable?\n1  2  3  4  5"),
        ("text", (1, 14), "'Was the lecture structure predictable?' with a Likert scale (1-5).")
    ]),
    ("Question 5 (Relevance)", [
        ("box", (1, 10), (8, 4), "How relevant was the lecture content to your current studies?\n1  2  3  4  5"),
        ("text", (1, 14), "'How relevant was the lecture content to your current studies?' with a Likert scale (1-5).")
    ]),
    ("Completion Screen", [
        ("box", (1, 10), (8, 4), "Thank you for your feedback!"),
        ("text", (1, 14), "'Thank you for your feedback!' with an option to close the survey.")
    ])
]

for ax, screen in zip(axs.flatten()[:8], screens):
    draw_screen(ax, screen[0], screen[1])
    plt.show()
fig.delaxes(axs[2, 2])

plt.tight_layout()
plt.show()
```
