```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv('A6 IS - Sheet1.csv')

# Group by method, intrusion, and object distance, then calculate mean and median
grouped = df.groupby(['method', 'intrusion', 'r'])['dmc'].agg(['mean', 'median'])

# Reset index to make method, intrusion, and r columns again
grouped = grouped.reset_index()

# Plotting
fig, ax = plt.subplots()

for method in grouped['method'].unique():
    for intrusion in grouped['intrusion'].unique():
        subset = grouped[(grouped['method'] == method) & (grouped['intrusion'] == intrusion)]
        ax.bar(subset['r'] + intrusion/100, subset['mean'], width=10, label=f'{method}-{intrusion}')

ax.set_xlabel('Object Distance (pixels)')
ax.set_ylabel('Mean Accuracy (pixels)')
ax.set_title('Mean Accuracy by Method and Object Distance')
ax.legend()

plt.show()
```
